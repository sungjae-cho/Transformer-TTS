```python
import synthesis
import os
import IPython.display as ipd

os.environ["CUDA_VISIBLE_DEVICES"]="4,5"
```

```python
class Args:
    def __init__(self, txt2mel_step=172000, mel2speech_step=100000, max_len=400):
        self.restore_step1 = txt2mel_step
        self.restore_step2 = mel2speech_step
        self.max_len = max_len
        self.sample_rate = 22050
        
    def get_txt2mel_step(self, txt2mel_step):
        self.restore_step1 = txt2mel_step
```

```python
args = Args(max_len=200)
```

# 1. The ouput at the last checkpoint

```python
args.get_txt2mel_step(574645)
text = "KAIST is a research institute."
wav = synthesis.synthesis(text, args)
ipd.Audio(wav, rate=args.sample_rate)
```
